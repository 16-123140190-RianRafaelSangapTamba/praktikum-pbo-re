```python
#Minesweeper Game

import random

# Ukuran papan permainan
dimensi = 3

# Membuat papan dengan tanda '?' yang menandakan kotak belum dibuka
papan_permainan = [['?' for _ in range(dimensi)] for _ in range(dimensi)]

# Menempatkan bom secara acak pada salah satu kotak
posisi_bom_bar = random.randint(0, dimensi - 1)
posisi_bom_kol = random.randint(0, dimensi - 1)

# Fungsi untuk menampilkan papan
def tampilkan_papan():
    for baris in papan_permainan:
        print(' '.join(baris))

# Fungsi untuk memeriksa apakah pemain menang
def cek_menang():
    for baris in papan_permainan:
        if '?' in baris:  # Jika masih ada kotak yang belum dibuka
            return False
    return True

# Fungsi utama untuk memulai permainan
def mulai_permainan():
    print("Selamat datang di Peta Tambang! Papan 3x3, satu bom tersembunyi.")

    while True:
        tampilkan_papan()  # Menampilkan papan yang diperbarui

        # Meminta input dari pemain untuk memilih kotak
        try:
            baris = int(input("Pilih nomor baris (1-3): ")) - 1  # Menyesuaikan input untuk indeks 0
            kolom = int(input("Pilih nomor kolom (1-3): ")) - 1  # Menyesuaikan input untuk indeks 0
        except ValueError:
            print("Masukkan angka yang valid.")
            continue

        if not (0 <= baris < dimensi and 0 <= kolom < dimensi):
            print("Input tidak valid. Pastikan baris dan kolom antara 1 dan 3.")
            continue

        if papan_permainan[baris][kolom] != '?':
            print("Kotak sudah dibuka. Coba pilih kotak lain.")
            continue

        # Jika kotak yang dipilih berisi bom
        if baris == posisi_bom_bar and kolom == posisi_bom_kol:
            papan_permainan[baris][kolom] = 'X'
            tampilkan_papan()
            print("Boom! Kamu kalah!")
            break
        else:
            papan_permainan[baris][kolom] = 'O'
            print("Anda aman, Lanjutkan!")  # Pesan jika pilihan bukan bom

        # Mengecek apakah pemain telah menang
        if cek_menang():
            tampilkan_papan()
            print("Selamat! Kamu menang!")
            break

if __name__ == "__main__":
    mulai_permainan()
```

```python
#Hangman Game

import random

# Daftar kata yang akan ditebak
daftar_kata = ["python", "programming", "hangman", "developer", "computer", "keyboard"]

# Gambar hangman
gambar_hangman = [
    """
       +---+
       |   |
           |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       üòü  |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       üòü  |
       |   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       üò∞  |
      /|   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       üò®  |
      /|\\  |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       üòµ  |
      /|\\  |
      /    |
           |
    =========
    """,
    """
       +---+
       |   |
       üíÄ  |
      /|\\  |
      / \\  |
           |
    =========
    """
]

# Fungsi untuk menampilkan kata yang sudah ditebak
def tampilkan_kata(kata, huruf_tebakan):
    hasil = ''
    for huruf in kata:
        if huruf in huruf_tebakan:
            hasil += huruf + ' '
        else:
            hasil += '_ '
    return hasil.strip()

# Fungsi utama permainan Hangman
def main():
    kata = random.choice(daftar_kata)
    huruf_tebakan = []
    kesempatan = 6

    print("üéÆ Selamat datang di dunia Hangman!")
    print("üî§ Aku telah memikirkan sebuah kata... bisakah kamu menebaknya?")
    print("Kamu punya 6 nyawa. Gunakan dengan bijak! üí°\n")

    while kesempatan > 0:
        print(gambar_hangman[6 - kesempatan])
        print("üìò Kata misterius:", tampilkan_kata(kata, huruf_tebakan))
        print("üìú Huruf yang sudah kamu coba:", ' '.join(huruf_tebakan))
        print(f"‚ù§Ô∏è Kesempatan tersisa: {kesempatan}")

        tebakan = input("üëâ Masukkan satu huruf: ").lower()

        if not tebakan.isalpha() or len(tebakan) != 1:
            print("üö´ Oops! Masukkan hanya *satu* huruf yang valid!\n")
            continue

        if tebakan in huruf_tebakan:
            print("‚ö†Ô∏è Kamu sudah menebak huruf itu sebelumnya. Coba yang lain!\n")
            continue

        huruf_tebakan.append(tebakan)

        if tebakan not in kata:
            print(f"üò¢ Oh tidak! '{tebakan}' tidak ada dalam kata.")
            kesempatan -= 1
        else:
            print(f"üéâ Yes! Huruf '{tebakan}' ada dalam kata!")

        if all(huruf in huruf_tebakan for huruf in kata):
            print("\nüèÜ SELAMAT! Kamu berhasil menebak kata:", kata.upper())
            print("üéä Kamu menyelamatkan si stickman dari tiang gantungan!")
            break
        print()  # Baris kosong untuk spasi

    else:
        print(gambar_hangman[6])
        print("üí• GAME OVER!")
        print("üò¢ Kamu gagal menyelamatkan stickman.")
        print("üîë Kata yang benar adalah:", kata.upper())

if __name__ == "__main__":
    main()
```
